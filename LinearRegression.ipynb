```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/Salary_Data.csv')
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
```

```python
# Conjunto de entrenamiento
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 1/3, random_state = 0 )
```

```python
# Creando el modelo y probando los datos de entrenamiento
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit( X_train, y_train )
```

```python
# LLevando los datos de testing al modelo, el metodo predict devuelve los datos de Y ya predecidos 
y_predict = regression.predict( X_test )
```

```python
# Visualizar los resultados de entrenamiento
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Sueldo vs Año de experiencia')
plt.xlabel('Años de experiencia')
plt.ylabel('Sueldo')
```

```python
# Visualizar el conjunto de testing
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, regression.predict(X_train) , color = 'blue')
plt.title('Experiencia vs Salario con datos de testing')
plt.xlabel(' Experiencia ')
plt.ylabel('Salario')
```
