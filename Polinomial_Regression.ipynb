```python
# Regresion polinomial
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/Position_Salaries.csv')
X = data.iloc[:, 1:2].values
Y = data.iloc[:, 2].values
```

```python
# Realizando la regresion lineal simple
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
# Ajustar el modelo a la regresion polinomicoa
from sklearn.preprocessing import PolynomialFeatures
# Como se desea un polinomio de grado 2 entonces se obtienen las variables predictoras x1 y x2^2, esto se hace con la libreria importada
poly_reg = PolynomialFeatures(4)
# Se obtienen la matriz donde cada columna es el valor de X desde grado 0 hasta el 2
X_poly = poly_reg.fit_transform( X )
# Se realiza la regresion con la misma instancia de linearregression
lin_reg_2 = LinearRegression()
# Creamos el modelo
lin_reg_2.fit( X_poly, Y )
```

```python
# Graficamos los resultados del modelo de regresion lineal
plt.scatter(X, Y, color = "red")
plt.plot(X, lin_reg.predict(X), color = "blue")
plt.title("Modelo de regresión lineal")
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo')
```

```python
# Modelo de regresion lineal polinomico
plt.scatter(X, Y, color = "red")
plt.plot(X, lin_reg_2.predict(X_poly), color = "blue")
plt.title("Modelo de regresión lineal polinomica")
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo')
```

```python
# Mejorando el modelo prediciendo un conjunto de valores
# Creamos un array con el valor minimo y maximo con intervalor de 0.1
X_grid = np.arange( min(X), max(X), 0.1)
# Como esto devuelve un vector entonces con la funcion reshape le cambiamos la dimension en donde le indicamos las filas y las columnas, 
# de esta forma se convierte en una matriz
X_grid = X_grid.reshape(len(X_grid), 1)

# Obtenemos los variables de la variable predicotra
X_poly = poly_reg.fit_transform( X_grid )
# Grafica
plt.scatter(X, Y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(X_poly), color = "blue")
plt.title("Modelo de regresión lineal polinomica")
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo')
```
